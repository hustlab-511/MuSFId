```python
# 滤波
import predictNet_use
# from Toolkit import *
import pandas as pd
import numpy as np
import torch
import matplotlib.pylab as plt

# 中值滤波降采样  1 1000
def avgpool(*,data,stride):
    length = data.shape[-1]
    ans = torch.zeros(1,1,int(length/stride)-2)
    for i in range(int(length/stride)-2):
        ans[0][0][i] = torch.sum(data[0][0][i*stride:i*stride+stride])/stride
    return ans

seed = 0
np.random.seed(seed)
torch.manual_seed(seed)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

begin = 7500
end = begin+25700
# end = begin+30000
data = np.array(pd.read_csv('/root/zqh/BCGOrigin_Data/kansas_bcg1/pa6.csv'))[begin:end,9:10]
# data = np.array(pd.read_csv('/root/zqh/Getian/BCG_hy_2023-08-30.csv'))[begin:end, 1:2]

# begin = 7500
# end = begin+5700
# data = np.array(pd.read_csv('/root/zqh/BCGOrigin_Data/ZQH040601bcg.csv'))[begin:end]
# data = np.array(pd.read_csv('/root/zqh/BCGOrigin_Data/ZQH0329.csv'))[begin:end]

data = data.reshape(1,1,end-begin)
data = data.astype(np.float32)
data = torch.from_numpy(data)
data = avgpool(data=data,stride=8) # Kansas数据预处理

# data = data.view(1,1,data.shape[-1]).cuda() # cuda
data = data.view(1,1,data.shape[-1]).to(device) # 非 cuda

# 去趋势
# deckernel_size = 101
deckernel_size = 51
avgpool1d = torch.nn.AvgPool1d(deckernel_size,stride=1)
T = avgpool1d(data) # 周期趋势
oral_data = data[:,:,deckernel_size//2 : data.shape[-1]-deckernel_size//2]
# oral_data = data
data = data[0][0][deckernel_size//2 : data.shape[-1]-deckernel_size//2] - T.squeeze(0).squeeze(0)
# data = data - T
data = data.view(1,1,data.shape[-1])
print(data.shape)

# 模型训练
epoch = 200
# model1 = predictNet.first_1_net().to(device=device) # cuda
model1 = predictNet_use.predictNet(kernel_size1=7, kernel_size2=101, data_length=data.shape[-1]) # 非 cuda
model1 = predictNet_use.trainPredictNet(model=model1,data=data,T=T,lr=0.0005,epoch=epoch)
torch.save(model1, '/root/zqh/Save_Model/SeriesDecompModel0104.pth')

output11, output22 = model1(data)
# output1, output2 = model1.returnAns(data)
data = data.to(device)
output11 = output11.to(device)
output22 = output22.to(device)
output = output11 + output22
print(output.shape)
showbegin = 1500
showend = 2500
plt.subplot(511)
plt.title("data")
plt.plot(oral_data[0][0][showbegin:showend])
plt.subplot(512)
plt.title("Trend")
plt.plot(T[0][0][showbegin:showend])
plt.subplot(513)
plt.title("output")
plt.plot(output[0][0][showbegin:showend].detach().numpy())
plt.subplot(514)
plt.title("small")
plt.plot(output11[0][0][showbegin:showend].detach().numpy())
plt.subplot(515)
plt.title("big")
plt.plot(output22[0][0][showbegin:showend].detach().numpy())
plt.show()

# 测试数据
# begin = 8000
# end = begin+5700
begin = 8000
end = begin+5700
# data = np.array(pd.read_csv('/root/zqh/Getian/2021hy/hy_bcg_ch2.csv'))[begin:end ,1:2]
# data = np.array(pd.read_csv('/root/zqh/BCGOrigin_Data/WCM040602bcg.csv'))[begin:end]
data = np.array(pd.read_csv('/root/zqh/BCGOrigin_Data/ZQH040601bcg.csv'))[begin:end]
# data = np.array(pd.read_csv('/root/zqh/BCGOrigin_Data/ZQH0329.csv'))[begin:end]
# data = np.array(pd.read_csv('/root/zqh/BCGOrigin_Data/kansas_bcg1/pa6.csv'))[begin:end,9:10]
data = data.reshape(1,1,end-begin)
data = data.astype(np.float32)
data = torch.from_numpy(data)
# data = avgpool(data=data,stride=8) # Kansas数据预处理

# data = data.view(1,1,data.shape[-1]).cuda() # cuda
data = data.view(1,1,data.shape[-1]).to(device) # 非 cuda
# 去趋势
# deckernel_size = 151
T = avgpool1d(data) # 周期趋势
oral_data = data[:,:,deckernel_size//2 : data.shape[-1]-deckernel_size//2]
# oral_data = data
data = data[0][0][deckernel_size//2 : data.shape[-1]-deckernel_size//2] - T.squeeze(0).squeeze(0)
# data = data - T
data = data.view(1,1,data.shape[-1])
print(data.shape)
# 输入模型得到结果
output11, output22 = model1(data)
# output1, output2 = model1.returnAns(data)
data = data.to(device)
output11 = output11.to(device)
output22 = output22.to(device)
output = output11 + output22
print(output.shape)

plt.subplot(411)
plt.title("origin data")
plt.plot(oral_data[0][0][showbegin:showend])
plt.subplot(412)
plt.title("Trend")
plt.plot(T[0][0][showbegin:showend])
# plt.subplot(513)
# plt.title("output")
# plt.plot(output[0][0][showbegin:showend].detach().numpy())
plt.subplot(413)
plt.title("small")
plt.plot(output11[0][0][showbegin:showend].detach().numpy())
plt.subplot(414)
plt.title("big")
plt.plot(output22[0][0][showbegin:showend].detach().numpy())
# plt.savefig('/root/zqh/EPS/seriesdecomp.eps', format='eps', dpi=1000)
plt.show()

# 保存eps图片
# f1 = plt.figure(figsize = (6, 2))
# plt.plot(oral_data[0][0][showbegin:showend])
# torch.save(oral_data[0][0][showbegin:showend], '/root/zqh/ImageData/oral_data.pth')
# # f1.savefig('/root/zqh/EPS/origin_data.eps', format='eps', dpi=1000)
# f1.savefig('/root/zqh/EPS/origin_data.eps')
# f2 = plt.figure(figsize = (6, 2))
# plt.plot(T[0][0][showbegin:showend])
# torch.save(T[0][0][showbegin:showend], '/root/zqh/ImageData/trend_data.pth')
# # f2.savefig('/root/zqh/EPS/Trend_data.eps', format='eps', dpi=1000)
# f2.savefig('/root/zqh/EPS/Trend_data.eps')
# f3 = plt.figure(figsize = (6, 2))
# plt.plot((oral_data-T)[0][0][showbegin:showend])
# # f3.savefig('/root/zqh/EPS/non_Trend_data.eps', format='eps', dpi=1000)
# f3.savefig('/root/zqh/EPS/non_Trend_data.eps')
# f4 = plt.figure(figsize = (6, 2))
# plt.plot(output11[0][0][showbegin:showend].detach().numpy())
# torch.save(output11[0][0][showbegin:showend].detach().numpy(), '/root/zqh/ImageData/small_data.pth')
# # f4.savefig('/root/zqh/EPS/Small_data.eps', format='eps', dpi=1000)
# f4.savefig('/root/zqh/EPS/Small_data.eps')
# f5 = plt.figure(figsize = (6, 2))
# plt.plot(output22[0][0][showbegin:showend].detach().numpy())
# torch.save(output22[0][0][showbegin:showend].detach().numpy(), '/root/zqh/ImageData/big_data.pth')
# # f5.savefig('/root/zqh/EPS/Big_data.eps', format='eps', dpi=1000)
# f5.savefig('/root/zqh/EPS/Big_data.eps')


for name, param in model1.named_parameters():
    print("Layer Name:", name)
    print("Parameter Shape:", param.shape)
```
